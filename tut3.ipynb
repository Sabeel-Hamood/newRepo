```python
import json
print("Dictionary: ", json.dumps({"name": "John", "age": 30}))
print("List:       ", json.dumps(["apple", "bananas"]))
print("Tuple:      ", json.dumps(("apple", "bananas")))
print("String:     ", json.dumps("hello"))
print("Integer:    ", json.dumps(42))
print("Float:      ", json.dumps(31.76))
print("True:       ", json.dumps(True))
print("False:      ", json.dumps(False))
print("None:       ", json.dumps(None))
```

```python
json_data = '{"acronym": "BLD", "name": "Boulder Colorado", \
              "latitude": 40.00, "longitude": -105.25}'
python_obj = json.loads(json_data)
print("Name:     ", python_obj["name"])
print("Acronym:  ", python_obj["acronym"])
print("Latitude: ", python_obj["latitude"])
print("Latitude: ", python_obj["longitude"])

print()
```

```python
import json
studentsList = []
print("Started Reading JSON file which contains multiple JSON document")
with open('students.txt') as f:
    for jsonObj in f:
        studentDict = json.loads(jsonObj)
        studentsList.append(studentDict)

print("Printing each JSON Decoded Object")
for student in studentsList:
    print(student["firstName"], student["lastName"], student["email"], student["Courses"],student["Link"])

with open('/content/drive/MyDrive/students1.json', "w") as fid: 
     json.dump(studentsList, fid)
```

```python
import pandas as pd
from ipywidgets import widgets
from IPython.display import display
from traitlets.utils import descriptions
#@title Default title text
first_name = "Sabeel" #@param ["Name","Sabeel", "Ehsan","Sara", "Safwan", "Bshara", "Mohamd"]
for student in studentsList:
  if(student["firstName"]==first_name):
      firstName = widgets.Label(value=str("First Name: "+student["firstName"]), description="First_Name:")
      last_name = widgets.Label(value=str("Last Name: "+student["lastName"]), description="Last_Name:")
      email = widgets.Label(value=str("Email: "+student["email"]), description='Email:')
      courses = widgets.Label(value=str(student["Courses"]), description='Courses:')
      display(firstName)
      print("")
      display(last_name)
      print("")
      display(email)
      print("")
      print("Courses has been taking by "+ first_name + " are: ")
      display(courses)
      print("")
      TVseries = widgets.Text(value=str(""),description='TV Series:')
      display(TVseries)
      btn = widgets.Button(description='Update')
      def button_eventhandler(obj):
          if(TVseries.value!=""):
            newData={"TVseries":TVseries.value}
            student.update(newData)
            i=0;
            for stud in studentsList:
              if(stud["firstName"]==first_name):
                  studentsList[i].update(newData)
                  print(studentsList[i])
              else:
                i+=1
      btn.on_click(button_eventhandler)
      print(""),  print("")
      display(btn)
with open('/content/drive/MyDrive/students1.json', "w") as fid: 
     json.dump(studentsList, fid)

```
